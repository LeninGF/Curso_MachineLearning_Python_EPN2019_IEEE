### Operaciones sobre im치genes
Leer la imagen descargada y detectar los bordes

```python
path_mesa = os.path.join(dir_actual, 'imagenes', 'risco.png')
img_mesa = imread(path_mesa)
plt.imshow(img_mesa)
```

```python
from skimage.color import rgb2gray
img_mesa_gris = rgb2gray(img_mesa)
plt.imshow(img_mesa_gris, cmap='gray')
```

```python
from skimage import feature
bordes = feature.canny(img_mesa_gris, sigma=0.3)
plt.imshow(bordes)
```
